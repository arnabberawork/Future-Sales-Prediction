# Problem Statement

### Description
Walmart, one of the leading retail stores in the US, aims to predict sales and demand accurately. Several events and holidays significantly impact daily sales. Walmart faces challenges such as unforeseen demand and stockouts due to inappropriate machine learning algorithms. An ideal machine learning model will accurately predict demand while accounting for factors such as economic conditions (e.g., CPI, Unemployment Index).

Walmart runs promotional markdown events throughout the year. These markdowns are implemented before major holidays such as Super Bowl, Labour Day, Thanksgiving, and Christmas.

Holiday weeks are weighted five times higher in evaluation compared to non-holiday weeks.
The challenge includes modeling the effects of markdowns on holiday weeks despite incomplete or non-ideal historical data.
Historical sales data for 45 Walmart stores located in different regions are provided for analysis.

### Dataset Information
This historical dataset covers sales data from 2010-02-05 to 2012-11-01, available in the file Walmart_Store_sales.

- Fields:
    1. Store - The store number.
    2. Date - The week of sales.
    3. Weekly_Sales - Sales for the given store during the week.
    4. Holiday_Flag - Indicates whether the week is a holiday week:
    1 – Holiday week
    0 – Non-holiday week
      
    5. Temperature - Temperature on the day of sales.
    6. Fuel_Price - Cost of fuel in the region.
    7. CPI - Consumer Price Index.
    8. Unemployment - Unemployment rate.
    9. Holiday Events and Dates:
    - Super Bowl
    
    12-Feb-2010, 11-Feb-2011, 10-Feb-2012, 8-Feb-2013
    - Labour Day
    
    10-Sep-2010, 9-Sep-2011, 7-Sep-2012, 6-Sep-2013
    - Thanksgiving
    
    26-Nov-2010, 25-Nov-2011, 23-Nov-2012, 29-Nov-2013
    - Christmas
    
    31-Dec-2010, 30-Dec-2011, 28-Dec-2012, 27-Dec-2013

Data Understanding and Cleanup:

Explore the dataset and perform necessary cleaning if required.

A single feature.
Multiple features.
Model Evaluation and Comparison:

Evaluate the performance of the models using metrics such as:
R² (R-squared)

Compare the scores of the respective models.

There are not so much outliers in the dataset, will consider upper and lower boundary 5%

```python
def remove_outliers(df, columns):
    for col in columns:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        # Calculate the Interquartile Range (IQR)
        IQR = Q3 - Q1
        # Define the upper and lower bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Filter the DataFrame to remove outliers
        df = df[ (df[col] >= lower_bound) & (df[col] <= upper_bound) ]
        print(f" Current Shape after removing outliers of {col}  : " , df.shape)
    return df
```

# Custom Functions

#### Evaluate Regression

```python
def evaluate_regression(X_true, y_true, estimator, data_name, dataset_type):
    """
    Evaluate a regression model using common metrics.

    Parameters:
        X_true: array-like
            The input features of the dataset to evaluate.
        y_true: array-like
            The true target values.
        estimator: object
            The trained regression model to evaluate.
        data_name: str
            The name of the dataset (for logging purposes).
        dataset_type: str
            Type of the dataset (e.g., 'Training', 'Test').

    Returns:
        A dictionary containing evaluation metrics.
    """
    print(f"Evaluating {dataset_type} Set: data set name - {data_name}")

    # Predictions
    y_pred = estimator.predict(X_true)

    # Mean Absolute Error (MAE)
    mae = mean_absolute_error(y_true, y_pred)
    print(f"Mean Absolute Error (MAE): {round(mae, 4)}")

    # Mean Squared Error (MSE)
    mse = mean_squared_error(y_true, y_pred)
    print(f"Mean Squared Error (MSE): {round(mse, 4)}")

    # Root Mean Squared Error (RMSE)
    rmse = np.sqrt(mse)
    print(f"Root Mean Squared Error (RMSE): {round(rmse, 4)}")

    # R-squared (R2) Score
    r2 = r2_score(y_true, y_pred)
    print(f"R-squared (R2) Score: {round(r2, 4)}")

    # Return all metrics as a dictionary
    metrics = {
        "Estimator": estimator,
        "Type": dataset_type,
        "Data name" : data_name,
        "MAE": mae,
        "MSE": mse,
        "RMSE": rmse,
        "R2": r2
    }

    return metrics
```

#### Run Model

```python
# function to grid search the best parameters for the model
def run_model(model,X_model_train,y_model_train,param_grid,scoring='neg_mean_squared_error',n_splits=5):
    # Log Start Time
    start_time=time.time()
    print("Execution Start Time : " , datetime.now())
    cv=KFold(n_splits=n_splits,shuffle=True,random_state=42)
    grid=GridSearchCV(model, param_grid=param_grid, cv=cv, scoring=scoring, verbose=1, n_jobs=-1)
    grid.fit(X_model_train,y_model_train)
    # Log End time
    end_time = time.time()
    print("Execution End Time : " , datetime.now())
    # Calculate the elapsed time
    execution_time = end_time - start_time
    print(f"Execution time: {execution_time} seconds")
    return grid.best_estimator_
```

#### Find Model Coefficients

```python
def get_sorted_coefficients(feature_names, coefficients, ascending=False):
    """
    Create a DataFrame pairing feature names with their coefficients and sort them.

    Parameters:
        feature_names (list or array-like): Names of the features.
        coefficients (list or array-like): Corresponding coefficients for the features.
        ascending (bool): Sort order, False for descending (default) and True for ascending.

    Returns:
        pd.DataFrame: A DataFrame containing feature names and coefficients, sorted by importance.
    """
    # Create a DataFrame
    coefficients_df = pd.DataFrame({
        "Feature": feature_names,
        "Coefficient": coefficients
    }).sort_values(by="Coefficient", ascending=ascending)

    return coefficients_df
```

## Custom wrapper for Keras model

```python
class KerasRegressorCustom(BaseEstimator, RegressorMixin):
    def __init__(self, build_fn=None, neurons=64, learning_rate=0.01, activation='relu', epochs=30, batch_size=32, dropout_rate=0.2):
        self.build_fn = build_fn
        self.neurons = neurons
        self.learning_rate = learning_rate
        self.activation = activation
        self.epochs = epochs
        self.batch_size = batch_size
        self.dropout_rate = dropout_rate

    def fit(self, X, y):
        # Build model using provided function
        self.model = self.build_fn(neurons=self.neurons, learning_rate=self.learning_rate, activation=self.activation, dropout_rate=self.dropout_rate)
        # Fit model
        self.model.fit(X, y, epochs=self.epochs, batch_size=self.batch_size, verbose=0)
        return self

    def predict(self, X):
        return self.model.predict(X)

    def score(self, X, y):
        # Compute MSE for scoring in GridSearchCV (negative because GridSearchCV minimizes the score)
        y_pred = self.predict(X)
        mse = np.mean((y - y_pred) ** 2)
        return -mse  # Negative MSE for compatibility with GridSearchCV scoring
```

## ANN

```python
def create_ann_model(learning_rate=0.01, neurons=64, activation='relu'):
    model = Sequential()
    model.add(Dense(neurons, input_dim=X_train.shape[1], activation=activation))  # Input layer
    model.add(Dense(neurons, activation=activation))  # Hidden layer
    model.add(Dense(1))  # Output layer (for regression)

    optimizer = Adam(learning_rate=learning_rate)
    model.compile(loss='mean_squared_error', optimizer=optimizer)
    return model
```

```python
params = {
    'neurons': [32, 64],
    'learning_rate': [0.001, 0.01],
    'batch_size': [16, 32],
    'epochs': [20, 30]
}
# Wrap the model with custom KerasRegressor
ann_model = KerasRegressorCustom(build_fn=create_ann_model)

# Run GridSearchCV for ANN
model = run_model(ann_model, X_train, y_train, params, scoring='neg_mean_squared_error', n_splits=3)

# Evaluate the model on train and test sets
train_results = evaluate_regression(X_train, y_train, estimator=model, data_name='X_train', dataset_type='ANN')
test_results = evaluate_regression(X_test, y_test, estimator=model, data_name='X_test', dataset_type='ANN')

# Convert dictionaries to DataFrames
test_results_df = pd.DataFrame([test_results])
# Combine the two DataFrames
model_summary = pd.concat([model_summary, test_results_df], ignore_index=True)
# Display the results DataFrame
print(model_summary)
# Save the best ANN model
pickle.dump(model, open("ann_model.pkl", "wb"))
```

## LSTM

```python
# LSTM Model Creation
def create_lstm_model(learning_rate=0.01, neurons=32, dropout_rate=0.2, activation='relu'):
    model = Sequential()
    model.add(LSTM(units=neurons, input_shape=(X_train.shape[1], 1), activation=activation, return_sequences=False))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1))  # Output layer (for regression)

    optimizer = Adam(learning_rate=learning_rate)
    model.compile(loss='mean_squared_error', optimizer=optimizer)
    return model
```

```python
# Reshape the input data for LSTM: (samples, time steps, features)
X_train_lstm = np.reshape(X_train.values, (X_train.shape[0], X_train.shape[1], 1))
X_test_lstm = np.reshape(X_test.values, (X_test.shape[0], X_test.shape[1], 1))

# Hyperparameter Grid for LSTM
params = {
    'neurons': [32],
    'learning_rate': [0.001, 0.01],
    'batch_size': [16],
    'epochs': [30]
}

# Wrap the model with custom KerasRegressor
lstm_model = KerasRegressorCustom(build_fn=create_lstm_model)

# Run GridSearchCV for LSTM
model = run_model(lstm_model, X_train_lstm, y_train, params, scoring='neg_mean_squared_error', n_splits=3)

# Evaluate the model on train and test sets
train_results = evaluate_regression(X_train_lstm, y_train, estimator=model, data_name='X_train_lstm', dataset_type='LSTM')
test_results = evaluate_regression(X_test_lstm, y_test, estimator=model, data_name='X_test_lstm', dataset_type='LSTM')

# Convert dictionaries to DataFrames
test_results_df = pd.DataFrame([test_results])
# Combine the two DataFrames
model_summary = pd.concat([model_summary, test_results_df], ignore_index=True)

# Display the results DataFrame
print(model_summary)
# Save the best LSTM model
pickle.dump(model, open("lstm_model.pkl", "wb"))
```

# Evalution and Conclusion

### Observations:
    - Linear Regression (LR), Ridge Regression, Lasso Regression, and XGBoost models show good performance with an R² score between 0.95 and 0.96, which indicates a strong fit to the data and effective prediction of future sales.
    - Polynomial Regression generated extreme values for the MAE, MSE, RMSE, and R² metrics, indicating that the model is severely overfitting the data or that its assumptions are not aligned with the nature of the problem. This suggests that polynomial features may not be ideal for sales forecasting in this case.
    - The Decision Tree Regressor and Random Forest Regressor show lower performance, with R² scores of 0.8325 and 0.8607, respectively. These models performed worse than linear models and XGBoost, which suggests they may not capture the data's underlying patterns as effectively.
    - The LSTM (Long Short-Term Memory) model, despite being a recurrent neural network typically suited for sequential data, performed poorly with a negative R² score of -3.3931. This indicates that the LSTM model is unable to capture the patterns in the sales data correctly, possibly due to insufficient data preprocessing, incorrect reshaping, or improper hyperparameter tuning. Additionally, LSTM models are generally more sensitive to training time, hyperparameters, and data sequences, which might need further adjustments.

### Conclusion:
    - XGBoost emerged as the best-performing model with the highest R² score of 0.9621, making it the most accurate for sales forecasting in this dataset. It outperformed all other models, including linear regression and neural networks.
    - Linear Regression models also performed well with an R² score around 0.95, indicating that simpler models can be effective for sales prediction when feature engineering is done correctly.
    - LSTM, despite its potential for time series prediction, did not perform well on this particular problem. Further adjustments, including fine-tuning the model architecture, hyperparameters, and feature engineering, might improve its performance.
    - Polynomial Regression should be avoided as it caused issues with overfitting and did not yield reliable results.

### Recommendations and Future Actions:
    - XGBoost should be considered the go-to model for future sales prediction tasks due to its robust performance and flexibility.
    - The LSTM model could be revisited with improved preprocessing and tuning of hyperparameters if there's a need for capturing sequential patterns or trends in the data over time.
    - Feature engineering, including adding temporal features or incorporating external factors like promotions or holidays, could further improve model accuracy.
